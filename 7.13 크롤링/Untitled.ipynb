```python
import requests
import lxml.html
```

```python
response = requests.get('https://www.hanbit.co.kr/store/books/new_book_list.html')
root = lxml.html.fromstring(response.content)
```

```python
for a in root.cssselect('.view_box a'):
    url = a.get('href')
    print(url)
```
